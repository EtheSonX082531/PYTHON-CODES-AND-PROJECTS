```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
      name:str
      age:int
      skills:list[str]
      final:str
```

```python
def greeting(state:AgentState)->AgentState:
       state["final"]=f"{state["name"]},Welcome to the system!"
       return state
```

```python
def user_age(state:AgentState)->AgentState:
      state["final"]=state["final"]+f"You are {state["age"]} years old!"
      return state
```

```python

def show_skills(state: AgentState) -> AgentState:
    skill_list = ""

    for el in state["skills"]:
        skill_list = skill_list + str(el) + ", "

    state["final"] = state["final"] + "You have skills in: " + skill_list
    return state
```

```python
graph=StateGraph(AgentState)
graph.add_node("greeting",greeting)
graph.add_node("user_age",user_age)
graph.add_node("show_skills",show_skills)
```

```python
graph.set_entry_point("greeting")
graph.add_edge("greeting","user_age")
graph.add_edge("user_age","show_skills")
graph.set_finish_point("show_skills")
```

```python
app=graph.compile()
output=app.invoke({"name":"Marian","age":22,"skills":["Cooking","Talking","Styling"]})

print(output["final"])
```
