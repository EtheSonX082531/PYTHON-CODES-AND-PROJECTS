```python
from typing import TypedDict,List
from langgraph.graph import StateGraph

class AgentState(TypedDict):
     values:List[int]
     name:str
     result:str

def process_values(state:AgentState)->AgentState:
    state["result"]=f"Hi {state["name"]},your given values sum is: {sum(state["values"])}"
    return state

graph=StateGraph(AgentState)

graph.add_node("prcs_val",process_values)
graph.set_entry_point("prcs_val")
graph.set_finish_point("prcs_val")
app=graph.compile()

output_state = app.invoke({"values":[1,2,3,4],"name":"Nila","result":""})

print(output_state["result"])
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph

class AgentState(TypedDict):
      values:list[int]
      name:str
      operation:str
      result:str

def sum_Or_Multiply(state:AgentState)->AgentState:
      if "+" in state["operation"]:
        ans=sum(state["values"])
      elif "*" in state["operation"]:
          ans=1
          for el in state["values"]:
                ans=ans*el
      else:
          ans="Invalid"
      state["result"]=f"Hi, {state["name"]}.Answer is {ans}."
      return state


graph=StateGraph(AgentState)

graph.add_node("sumOrMul",sum_Or_Multiply)
graph.set_entry_point("sumOrMul")
graph.set_finish_point("sumOrMul")
app=graph.compile()

output=app.invoke({"values":[1,2,3,4,5],"name":"Nila","operation":"*"})
print(output["result"])

output=app.invoke({"values":[1,2,3,4,5],"name":"Anik","operation":"+"})
print(output["result"])
```
