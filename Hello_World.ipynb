```python
from typing import TypedDict,Dict
from langgraph.graph import StateGraph

class AgentState(TypedDict):
      message : str

def greeting_node(state:AgentState)->AgentState:
    state["message"]="Hello "+state["message"]+",Good Morning❤️.."
    return state

graph=StateGraph(AgentState)

graph.add_node("greeter",greeting_node)
graph.set_entry_point("greeter")
graph.set_finish_point("greeter")

app = graph.compile()

result = app.invoke({"message":"Mithila"})

print(result["message"])
```

```python
from langgraph.graph import StateGraph

input={"string":""}

def compliment_node(state:dict)->dict:
    compliment=" ,You are doing a great job by learning Langgraph!"
    state["string"]=state["string"]+compliment
    return state

graph=StateGraph(dict)

graph.add_node("compliment_",compliment_node)

graph.set_entry_point("compliment_")
graph.set_finish_point("compliment_")

program=graph.compile()

output=program.invoke({"string":"Anik"})

print(output["string"])
```
